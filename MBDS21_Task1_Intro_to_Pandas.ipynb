How to select multiple rows and columns from a pandas DataFrame?

```python
ufo.head(3)
#pick first 3 rows
```

```python
# .loc is for selecting things by label
# pd.loc[row,col]
ufo.loc[0,:]
```

```python
#pass in the list for multiple rows
ufo.loc[[0,1,3,5,10],:]
```

```python
#pass in the range for multiple continuous rows
ufo.loc[0:4,:]
```
