```python
import pandas as pd
import seaborn as sns
sns.__version__
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#df = pd.read_csv('USvideos.csv')
df = pd.read_csv("https://raw.githubusercontent.com/asukul/DS201/master/datasets/USvideos.csv")
```

```python
df.info()
```

```python
df.channel_title = df.channel_title.dropna()
```

```python
df.info()
```

```python
df.head(20)
```

```python
#shows top 5 likes
df.nlargest(5, 'likes')
```

```python
#shows top 10 likes and views
df.nlargest(10, ['likes','views'])
```

```python
sns.pairplot(df)
#pairplot will display a scatter plot of  a combination values columns, not a categoical columns
```

```python
sns.color_palette("deep")
```

```python
sns.pairplot(df[["category_id","views","likes"]], hue="category_id")
```

```python
sns.pairplot(df, vars=["views","likes"], hue="category_id", palette="deep")
```

```python
#select top 100 videos by top views 
df_top100views = df.nlargest(100, 'views')
```

```python
df_top100views.info()
```

```python
sns.pairplot(df_top100views[["category_id","views","likes"]], hue="category_id")
```

```python
sns.pairplot(df_top100views, vars=["category_id","views","likes"], hue="category_id", diag_kind="hist", palette="deep")
```
